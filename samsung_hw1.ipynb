### Introduction
Access to electricity is an important factor that supports effective learning environments in schools. Electricity enables lighting, digital tools, and other educational resources that improve the quality of education. At the same time, expanding electricity access is a key objective of global energy development.

This project explores a dataset on electricity access in schools and examines its relevance to Sustainable Development Goal 4 (Quality Education) and Sustainable Development Goal 7 (Affordable and Clean Energy). Through data visualization, the project aims to understand how electricity availability in schools relates to these two goals and evaluate whether the dataset can provide meaningful insights for analyzing progress toward them.

```python
# Her ülke için en güncel yılı seç
latest_df = df.sort_values("Year").groupby("Country_Name", as_index=False).tail(1)

# En yüksek ve en düşük 10 ülke
top_10 = latest_df.nlargest(10, "Electricity_Access_Percent").sort_values("Electricity_Access_Percent", ascending=True)
bottom_10 = latest_df.nsmallest(10, "Electricity_Access_Percent").sort_values("Electricity_Access_Percent", ascending=True)

# En fazla erişimi olan ülkeler grafiği
fig_top = px.bar(
    top_10,
    x="Electricity_Access_Percent",
    y="Country_Name",
    orientation="h",
    text="Electricity_Access_Percent",
    title="Elektriğe Erişimi En Fazla Olan 10 Ülke (En Güncel Yıla Göre)",
    hover_data=["Year", "ISO_Code"]
)

fig_top.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig_top.update_layout(
    xaxis_title="Elektrik Erişim Oranı (%)",
    yaxis_title="Ülke",
    template="plotly_white"
)
fig_top.show()

# En az erişimi olan ülkeler grafiği
fig_bottom = px.bar(
    bottom_10,
    x="Electricity_Access_Percent",
    y="Country_Name",
    orientation="h",
    text="Electricity_Access_Percent",
    title="Elektriğe Erişimi En Az Olan 10 Ülke (En Güncel Yıla Göre)",
    hover_data=["Year", "ISO_Code"]
)

fig_bottom.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig_bottom.update_layout(
    xaxis_title="Elektrik Erişim Oranı (%)",
    yaxis_title="Ülke",
    template="plotly_white"
)
fig_bottom.show()
```

```python
country_avg = (
    df.groupby("Country_Name")["Electricity_Access_Percent"]
    .mean()
    .reset_index()
    .sort_values(by="Electricity_Access_Percent", ascending=False)
)

top10 = country_avg.head(10)
bottom10 = country_avg.tail(10)

overall_mean = country_avg["Electricity_Access_Percent"].mean()
top_country = country_avg.iloc[0]
bottom_country = country_avg.iloc[-1]

print("General Insights:")
print(f"The overall average electricity access in schools across countries is {overall_mean:.2f}%.")
print(f"The country with the highest average access is {top_country['Country_Name']} with {top_country['Electricity_Access_Percent']:.2f}%.")
print(f"The country with the lowest average access is {bottom_country['Country_Name']} with {bottom_country['Electricity_Access_Percent']:.2f}%.")
print(f"The average of the top 10 countries is {top10['Electricity_Access_Percent'].mean():.2f}%.")
print(f"The average of the bottom 10 countries is {bottom10['Electricity_Access_Percent'].mean():.2f}%.")
```
